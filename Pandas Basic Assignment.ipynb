#### Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

In Pandas, a DataFrame is a two-dimensional, size-mutable, and heterogeneous tabular data structure with labeled axes (rows and columns). It can be thought of as a table of data, where each column can have a different data type. A DataFrame is similar to a spreadsheet or a SQL table and is one of the most commonly used data structures in Pandas.

On the other hand, a Series is a one-dimensional labeled array that can hold data of any type (integers, strings, floating-point numbers, Python objects, etc.). It's like a column in a DataFrame or a single column of data in a spreadsheet.

Example:

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Claire'],
    'Age': [25, 30, 27],
    'Gender': ['Female', 'Male', 'Female']
}
df = pd.DataFrame(data)
df
```

```python
type(df)
```

```python
import pandas as pd
data = [4,8,15,16,23,43]
series = pd.Series(data)
series
```

#### Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?

Pandas provides a wide range of functions to manipulate data in a DataFrame. Here are some common functions:

```python
import pandas as pd
df = pd.read_csv("C:/Users/alisa/OneDrive/Desktop/skills/1.Python/Week 8 (ML Toolbox)-Pandas/Titanic.csv")
```

1. head() and tail(): These functions allow you to view the first few or last few rows of a DataFrame, respectively. Useful for a quick look at the data. for Example:

```python
# Display the first five rows of the dataFrame
df.head()
```

```python
# Display the last 3 rows of the dataframe
df.tail(3)
```

2. describe(): Generates descriptive statistics of the numerical columns in the DataFrame, like mean, standard deviation, minimum, maximum, quartiles, etc.

```python
df.describe()
```

3. info(): Provides a concise summary of the DataFrame, including data types and non-null counts.

```python
df.info()
```

4. groupby(): Used to group data based on one or more columns and apply aggregate functions to the groups.

```python
df.groupby(['sex'])['age'].mean()
```

5. sort_values(): Sorts the DataFrame by specified columns.

```python
df.sort_values('name')
```

6. drop(): Used to remove rows or columns from the DataFrame.

```python
df.drop('sibsp',axis=1, inplace=True)
```

```python
df.head()
```

7. fillna(): Fills missing values in the DataFrame with a specified value or strategy.

```python
df['body'].fillna(df['body'].mean(), inplace=True)
```

```python
df.head()
```

8. apply(): Applies a function to each element or row/column in the DataFrame.

```python
df['age'] = df['age'].apply(lambda x: x+1)
```

```python
df
```

#### Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

In Pandas, both Series and DataFrame are mutable in nature, while the Panel is not.

Series: A Series is a one-dimensional labeled array, and you can modify its elements, add new elements, or change its properties.

DataFrame: A DataFrame is a two-dimensional tabular data structure. It's mutable, meaning you can add or remove columns, modify data in existing columns, and perform various operations to manipulate the DataFrame.

Panel: A Panel used to be a three-dimensional data structure in Pandas.

#### Q7. Create a DataFrame using multiple Series. Explain with an example.

```python
import pandas as pd

# Creating Series 
names = pd.Series(['Alice', 'Bob', 'Claire'])
ages = pd.Series([25, 30, 27])
genders = pd.Series(['Female', 'Male', 'Female'])

# Creating a dictionary of Series
data = {
    'Name': names,
    'Age': ages,
    'Gender': genders
}

# Creating a DataFrame from the dictionary
df = pd.DataFrame(data)
df
```
