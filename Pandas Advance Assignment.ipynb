#### Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row.

```python
import pandas as pd

def reindex_with_increment(df):
    new_index = pd.Index(range(1, len(df)*2, 2))  # Create a new index starting from 1, incrementing by 2
    df = df.set_index(new_index)
    return df

# Re-index the DataFrame
new_df = reindex_with_increment(df)
new_df
```

#### Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The function should print the sum to the console.

```python
import pandas as pd

def cal_sum_of_first_three(df):
    get_column = df['Values']
    sum_of_first_three = sum(get_column[:3])
    return sum_of_first_three

data = {
    'Values':[10, 20, 30, 40, 50]
}
df = pd.DataFrame(data)
result = cal_sum_of_first_three(df)
print(f"Sum of the first three values : ",result)
```

#### Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email addresses in the format 'username@domain.com'. Write a Python function that creates a new column 'Username' in df that contains only the username part of each email address.

```python
import pandas as pd

def extract_username(email):
    username = email.split('@')[0]  # Split the email at '@' and take the first part
    return username

def add_username_column(df):
    df['Username'] = df['Email'].apply(extract_username)
    return df

data = {
    'Email':['john.doe@example.com','ravindrajd23@gmail.com','tulsidaskhan@gmail.com']
}
df = pd.DataFrame(data)

df_with_username = add_username_column(df)
df_with_username
```

#### Q8. You have a Pandas DataFrame df with columns 'A', 'B', and 'C'. Write a Python function that selects all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10. The function should return a new DataFrame that contains only the selected rows.

```python
import pandas as pd

def select_rows(df):
    selected_rows = df[(df['A'] > 5) & (df['B'] < 10)]
    return selected_rows

# Sample DataFrame
data = {'A': [3, 8, 6, 2, 9],
        'B': [5, 2, 9, 3, 1],
        'C': [1, 7, 4, 5, 2]}
df = pd.DataFrame(data)
```

```python
df
```

```python
# Select rows based on conditions
selected_df = select_rows(df)
print(selected_df)
```

#### Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean, median, and standard deviation of the values in the 'Values' column.

```python
import pandas as pd

def calculate_stats(df):
    mean_value = df['Value'].mean()
    median_value = df['Value'].median()
    std_value = df['Value'].std()
    return mean_value,median_value,std_value

data = {
    'Value':[10,30,50,70,300]
}
df = pd.DataFrame(data)

mean,median,std = calculate_stats(df)
print(f"mean : {mean}")
print(f"median : {median}")
print(f"standard_deviation : {std}")
```

#### Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

```python
import pandas as pd

def select_rows_in_data_range(df,start_date,end_date):
    n = (df['Date'] >= start_date) & (df['Date'] <= end_date)
    selected_rows = df[n]
    return selected_rows
    
data = {
    'Date': pd.to_datetime(['2023-01-15', '2023-02-05', '2023-01-25', '2023-01-10'])
}
df = pd.DataFrame(data)

start_date = '2023-01-01'
end_date = '2023-01-31'
selected = select_rows_in_data_range(df,start_date,end_date)
selected
```
