#### Q2. How did you handle missing data in the wine quality data set during the feature engineering process? Discuss the advantages and disadvantages of different imputation techniques.

In the context of the red wine quality dataset, there were no missing values, as observed using df.isnull().sum(). However, in cases where missing data is present, several imputation techniques can be employed:

Imputation Techniques:
Mean/Median/Mode Imputation:

Replace missing values with the mean, median, or mode of the column.

Advantages: Simple and quick; works well with small amounts of missing data.

Disadvantages: Can distort data distribution, especially if many values are missing or if the data is skewed.

Forward/Backward Fill:

Replace missing values with the previous (forward fill) or next (backward fill) values in the sequence.

Advantages: Useful for time series data where trends exist; maintains continuity.

Disadvantages: Can propagate errors if missing data is systematic or if the trend changes significantly.

3. Encoding Categorical Variables:
   - Applied one-hot encoding to transform categorical variables (`gender`, `race_ethnicity`, `parental_level_of_education`, etc.) into numerical form for model compatibility.

4.Normalization:
   - Applied normalization techniques like Min-Max Scaling to ensure that all numerical features (like `average`) were on a similar scale, improving model performance.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot histograms for each feature
plt.figure(figsize=(20, 15))
for i, column in enumerate(df.columns):
    plt.subplot(4, 3, i + 1)
    sns.histplot(df[column], kde=True)
    plt.title(column)
plt.tight_layout()
plt.show()
```
