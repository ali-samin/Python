#### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.


Seaborn is a powerful Python data visualization library that provides a variety of plots to help you explore and communicate your data effectively. Here are some common plots you can create using Seaborn:

1. Scatter Plot

Use: Display the relationship between two continuous variables.

Function: sns.scatterplot()

2. Histogram

Use: Visualize the distribution of a single variable.
    
Function: sns.histplot()

3. Pair Plot

Use: Create scatterplots and histograms for multiple variables in a dataset to visualize pairwise relationships.
    
Function: sns.pairplot()

4. Joint Plot

Use: Combine scatter plots and histograms to visualize the joint distribution between two variables.
    
Function: sns.jointplot()

5. Bar Plot

Use: Compare categories or discrete variables by displaying their values as bars.
    
Function: sns.barplot()
