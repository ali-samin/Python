#### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations in various formats, such as charts, plots, graphs, and more. It provides a flexible and powerful way to visualize data and is widely used in data analysis, scientific research, and data visualization tasks.

Matplotlib's Pyplot module is a sub-library within Matplotlib that provides a simple and convenient interface for creating basic plots and visualizations. Here are five types of plots that can be created using the Pyplot module of Matplotlib:

1. Line Plot: Line plots are used to display data points connected by straight lines. They are commonly used to show trends or changes in data over a continuous or discrete interval

2. Scatter Plot: Scatter plots are used to visualize individual data points as dots on a 2D plane. They are often used to explore relationships between two variables.

3. Bar Plot: Bar plots (or bar charts) are used to represent categorical data with rectangular bars. They are useful for comparing values across different categories.

4. Histogram: Histograms are used to visualize the distribution of a continuous variable. They display data as bins or bars, with each bar representing a range of values.

5. Pie Chart: Pie charts are used to represent the composition of a whole by dividing it into slices, where each slice represents a proportion of the whole.

#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("This is X axis")
plt.ylabel("This is Y axis")
plt.title("Scatter Plot")
plt.show()
```

#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

he subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to organize multiple plots or visualizations in a grid layout, making it easier to compare and analyze different aspects of your data simultaneously. The subplot() function is particularly useful when you want to display multiple plots in the same figure, each representing different views, data sets, or aspects of your data.

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(x,y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])

plt.subplot(1, 2, 1)
plt.plot(x, y, 'b')  # 'b' specifies a blue line
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Subplot 1: Line Plot')


x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(1, 2, 2)
plt.plot(x, y, 'b')  # 'b' specifies a blue line
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Subplot 2: Line Plot')


# Display the subplots
plt.show()
```

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


A bar plot, also known as a bar chart or bar graph, is a type of data visualization that represents categorical data with rectangular bars. Each bar's length or height is proportional to the value it represents, making it a useful tool for comparing data across different categories. Bar plots are typically used for the following purposes:

Comparison: Bar plots make it easy to compare the quantities or frequencies of different categories or groups.

Ranking: They can be used to rank categories based on their values.

Distribution: Bar plots can also show the distribution of a categorical variable.

Part-to-Whole Relationship: They are useful for illustrating the contribution of each category to the whole.

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Bar plot")
```

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Bar plot")
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It is used to visualize the summary statistics of a dataset, including the median, quartiles, and potential outliers. Box plots are particularly useful for the following purposes:

Summary of Data Distribution: Box plots provide a concise summary of the central tendency, spread, and skewness of a dataset.

Identification of Outliers: They can help identify potential outliers in the data, which are data points that fall significantly outside the main distribution.

Comparison of Distributions: Box plots are useful for comparing the distributions of multiple datasets or groups, making them suitable for statistical analysis.

Visualization of Data Spread: The "whiskers" in a box plot show the range of data points within 1.5 times the interquartile range (IQR), giving an idea of how data is spread.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.boxplot([box1, box2], vert=False, labels=['Box 1', 'Box 2'], patch_artist=True, boxprops={'facecolor': 'lightblue'})
plt.xlabel('Values')
plt.title('Box Plot for box1 and box2')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```
